# Xgboost Regression

```python
dmatrix=xgb.DMatrix(data=x,label=y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=123)
```

```python
x.shape
```

```python
y.shape
```

```python
xgbModel = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, slient=False, alpha = 10, n_estimators = 100,)
```

```python
xgbModel.fit(x_train,y_train)

```

```python
preds = xgbModel.predict(x_test)
preds
```

```python
residual_xgb=preds-y_test
residual_xgb
```

```python
xgbplot = pd.DataFrame({'Predictied Values': list(preds), 'Residuals': list(residual_xgb)}, columns=['Predictied Values', 'Residuals'])
ax = sns.scatterplot(x='Predictied Values', y='Residuals', hue=xgbplot.columns,
                     data=xgbplot)
```

```python
xgbModel.feature_importances_
```

```python
# rmse between Y_test & Y_predict
rmse_xgb = np.sqrt(sklearn.metrics.mean_squared_error(y_test, preds))
rmse_xgb
```

```python
sklearn.metrics.r2_score(y_test, preds)
```

```python
print(sklearn.metrics.r2_score(y_test,preds)
      ,sklearn.metrics.mean_absolute_error(y_test,preds))
```
